```python
import numpy as np
import scipy as sp
import scipy.stats
import matplotlib
import matplotlib.pyplot as plt
import csv
import os.path
```

```python
def preprocess_file(file_name):
    out_file_name = file_name + '.csv'
#    if os.path.exists(out_file_name):
#        return
    with open(out_file_name, 'w') as outfile, open(file_name, 'r', encoding='utf-8') as infile:
        for line in infile:
            line = line.strip()
            if line.endswith(';'):
                outfile.write(line[0:-1] + '\n')
```

```python
def read_values(file_name, column_num):
    values = []
    with open(file_name,'r') as csvfile:
        rows = csv.reader(csvfile, delimiter=',')
        for row in rows:
            values.append(float(row[column_num]))
    return values
```

```python
plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.figsize'] = (16, 8)
plt.rcParams.update({'font.family': 'monospace', 'font.size': 16, 'lines.markersize': 16, 'legend.frameon': True})
```

```python
preprocess_file('results/exp1_2_percent_loss/exp1_logfile')
exp1_throughput = read_values('results/exp1_2_percent_loss/exp1_logfile.csv', 1)
exp1_send_rate = read_values('results/exp1_2_percent_loss/exp1_logfile.csv', 2)
x = np.linspace(0, 90, len(exp1_send_rate))
```

```python
fig, ax = plt.subplots()
ax.step(x=x, y=exp1_send_rate, label='send_rate')
ax.set(xlabel="delta [s]", ylabel="send rate [mbits]")
ax.legend(loc='upper right')
#plt.show()
plt.savefig("exp1_send_rate.pdf", bbox_inches='tight')
```

```python
fig, ax = plt.subplots()
ax.step(x=x, y=exp1_throughput, label='throughput')
ax.set(xlabel="delta [s]", ylabel="throughput [mbits]")
ax.legend(loc='upper right')
#plt.show()
plt.savefig("exp1_throughput.pdf", bbox_inches='tight')
```
