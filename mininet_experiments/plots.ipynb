```python
import numpy as np
import scipy as sp
import scipy.stats
import matplotlib
import matplotlib.pyplot as plt
import csv
```

```python
def read_values(file_name):
    with open(file_name,'r') as csvfile:
        rows = csv.reader(csvfile, delimiter=',')
        for row in rows:
            y = np.array(row)
            return y.astype(np.float)
```

```python
plt.rcParams['figure.figsize'] = (10, 6)

y = read_values('2_3_3.csv')
x = np.linspace(0.1, 1.0, 10)
plt.scatter(x, y, label='channelUtilization')
plt.xlabel('fullLoad')
plt.ylabel('channelUtilization')
plt.title('Task 2.3.3')
plt.legend(loc='upper left')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10, 6)

numHosts = 20
maxBackOffValues = range(1, 20 + 1)

y = read_values('4_1_2.csv')
x = maxBackOffValues
plt.scatter(x, y, label='channelUtilization')
plt.xticks(x)
plt.xlabel('maxBackOff')
plt.ylabel('channelUtilization')
plt.title('Task 4.1.2')
plt.legend(loc='upper left')
plt.show()
```

```python
samples = read_values('4_2_1/exp1.csv')
samples = samples.reshape((10, 30)).T

alpha = 0.05
n = 30
mean = np.mean(samples, axis=0)
var = np.var(samples, axis=0, ddof=1)
h = np.sqrt(var/n)*scipy.stats.t.ppf(1-alpha/2., n-1)

plt.rcParams['figure.figsize'] = (10, 10)

x = np.linspace(0.1, 1.0, 10)
plt.errorbar(x, mean, yerr=h, fmt='none', label='confidence interval')
plt.xlabel('fullLoad')
plt.ylabel('channelUtilization')
plt.title('Task 4.2.1 experiment 1')
plt.legend(loc='upper left')
plt.show()
```
